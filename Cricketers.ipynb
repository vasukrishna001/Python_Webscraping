```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.espncricinfo.com/cricketers/team/india-6"
page = requests.get(url)
soup= BeautifulSoup(page.text, "html")
print(soup)
```

```python
players_list = soup.find('div',class_="ds-grid lg:ds-grid-cols-2")
print(players_list)
```

```python
players_list_adj = soup.find_all('div',class_="ds-flex ds-flex-col")
print(players_list_adj)
```

```python
# Parse the title, age, and href for each player
list_player = []
for content in players_list_adj:
    # Extract the title (player's name)
    title = content.find('a', class_="ds-inline-flex ds-items-start ds-leading-none").text.strip()

    # Extract the hyperlink (href attribute)
    href = content.find('a', class_="ds-inline-flex ds-items-start ds-leading-none")['href']

    # Extract the age
    age = content.find('span', class_="ds-text-tight-m ds-font-regular ds-text-typo-mid3").text.strip()

    # Combine extracted data
    list_player.append({'name': title, 'link': href, 'age': age})

# Print the result
print(list_player)
```

```python
import pandas as pd
import re
# Define the base URL for hyperlinks
base_url = "https://www.espncricinfo.com"

# Initialize an empty list to store processed data
processed_data = []

# Loop through each player dictionary in the list
for player in list_player:
    # Extract details
    player_name = player['name']
    
    # Extract and format the age using regex to remove 'Age:' and just keep the number and unit (e.g., '22y, 164d')
    age_match = re.match(r"Age: (\d+y) (\d+d)", player['age'])
    if age_match:
        player_age = f"{age_match.group(1)},{age_match.group(2)}"  # Format as '22y,164d'
    else:
        player_age = "Unknown"
    
    player_link = base_url + player['link']  # Combine base URL with the hyperlink
    
    # Format the link as an Excel-friendly hyperlink
    excel_hyperlink = f'=HYPERLINK("{player_link}", "Profile")'
    
    # Append processed player data as a dictionary to the processed_data list
    processed_data.append({
        'PlayerName': player_name,
        'Age': player_age,
        'ProfileLink': excel_hyperlink
    })

# Create a DataFrame from the processed data
df = pd.DataFrame(processed_data)

# Save the DataFrame to a CSV file
output_path = "players_data.csv"
df.to_csv(output_path, index=False)

# Print the DataFrame
print(df)
```

```python
#Styling and Colouring Excel direrctly by using openpyxl with an excel workbook

import pandas as pd
import re
from openpyxl import Workbook
from openpyxl.styles import Font, Color, PatternFill
from openpyxl.utils.dataframe import dataframe_to_rows

# Define the base URL for hyperlinks
base_url = "https://www.espncricinfo.com"

# Initialize an empty list to store processed data
processed_data = []

# Loop through each player dictionary in the list
for player in list_player:
    # Extract details
    player_name = player['name']
    
    # Extract and format the age using regex to remove 'Age:' and just keep the number and unit (e.g., '22y, 164d')
    age_match = re.match(r"Age: (\d+y) (\d+d)", player['age'])
    if age_match:
        player_age = f"{age_match.group(1)},{age_match.group(2)}"  # Format as '22y,164d'
    else:
        player_age = "Unknown"
    
    player_link = base_url + player['link']  # Combine base URL with the hyperlink
    
    # Format the link as an Excel-friendly hyperlink
    excel_hyperlink = f'=HYPERLINK("{player_link}", "Profile")'
    
    # Append processed player data as a dictionary to the processed_data list
    processed_data.append({
        'PlayerName': player_name,
        'Age': player_age,
        'ProfileLink': excel_hyperlink
    })

# Create a DataFrame from the processed data
df = pd.DataFrame(processed_data)

# Create a Workbook and a Worksheet
wb = Workbook()
ws = wb.active
ws.title = "Players Data"

# Write DataFrame to Excel
for row in dataframe_to_rows(df, index=False, header=True):
    ws.append(row)

# Apply styling for the headers (Bold font)
header_font = Font(bold=True)
for cell in ws[1]:
    cell.font = header_font

# Apply styling for the hyperlinks (Blue color)
for row in ws.iter_rows(min_row=2, min_col=3, max_col=3):  # Column 3 is for ProfileLink
    for cell in row:
        cell.font = Font(color="0000FF", underline="single")  # Blue color and underlined

# Save the workbook to an Excel file
output_path = "players_data_with_hyperlinks.xlsx"
wb.save(output_path)
print(df)
print("Excel file saved with hyperlinks and styling.")
```
